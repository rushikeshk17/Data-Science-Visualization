```python
y1 = list(data2[data2['Sector']=='Rural']['General index'])
y1 = [x for x in y1 if str(x) != 'nan']
y2 = list(data2[data2['Sector']=='Urban']['General index'])
y2 = [x for x in y2 if str(x) != 'nan']
y3 = list(data2[data2['Sector']=='Rural+Urban']['General index'])
y3 = [x for x in y3 if str(x) != 'nan']
y = [y1,y2,y3]
```

## Line plot 

```python
df = data2[data2['Year']==2021]
df.reset_index(drop = True)
```

```python
x = df[df['Sector']=='Rural']['Month']
y = df[df['Sector']=='Rural']['General index']
x2 = df[df['Sector']=='Urban']['Month']
y2 = df[df['Sector']=='Urban']['General index']
x3 = df[df['Sector']=='Rural+Urban']['Month']
y3 = df[df['Sector']=='Rural+Urban']['General index']
```

```python
plt.figure(figsize = (10,7))
plt.plot(x,y,'go-', label='Rural', linewidth=2)
plt.plot(x2,y2,'r*-', label='Urban', linewidth=2)
plt.plot(x3,y3,'bx-', label='Urban', linewidth=2)
plt.xlabel('Month',size = 15)
plt.ylabel('Index',size = 15)
plt.title('General Consumer Price Index vs Month of Year',size = 15)
plt.legend()
plt.grid()
plt.show()
```
